```python
from huggingface_hub import hf_hub_download

# Download the large model files
hf_hub_download(repo_id="suno/bark", filename="pytorch_model.bin")
hf_hub_download(repo_id="suno/bark", filename="text_2.pt")
```

```python
# Install required packages
# !pip install TTS scipy --upgrade

# Import necessary libraries
from TTS.tts.configs.bark_config import BarkConfig
from TTS.tts.models.bark import Bark
from scipy.io.wavfile import write as write_wav

# Load the model configuration
config = BarkConfig()
model = Bark.init_from_config(config)

# Load the model checkpoint
model.load_checkpoint(config, checkpoint_dir="/content/bark/")  # Ensure this path is correct
model.to("cuda")

# Text to synthesize
text = "Hi everyone, it's Priyanka chopra! I’m excited to meet Shivam your are very good person , I feel glad to meet you "

# Generate audio
output_dict = model.synthesize(
    text,
    config,
    speaker_id=None,  # Ensure this is None if not using specific speaker IDs
    voice_dirs="/content/bark_voices",  # Ensure this is the correct path
    temperature=0.95
)

# Save the generated audio to a file
write_wav("output_pc.wav", 24000, output_dict["wav"])  # Use the correct sample rate

# Download the output file
from google.colab import files
files.download("output_pc.wav")
```
